## Removing Symbols

```python
print('Before Cleaning: ', ds['Last_Name'])
ds['Last_Name'] = ds['Last_Name'].str.strip("1234567890!@#$%^&*()+{}[]''\|,/<>>?._-")
print('After Cleaning: ', ds['Last_Name'])
```

## Phone Number Cleaning & Formating Symbols

```python

#ðŸ“Œ Store as a string (str), not as an integer.
#ðŸ“Œ Choose a format (555568970 or 555-456-2345) based on your requirements

print('Before Cleaning: ', ds['Phone_Number'])

# Step 0: Store as a string
ds['Phone_Number'] = ds['Phone_Number'].apply(lambda x: str(x))

# Step 1: Remove all non-numeric characters (but keep NaNs intact)
ds['Phone_Number'] = ds['Phone_Number'].str.replace(r'\D', '', regex=True)

# Step 2: Ensure only valid 10-digit numbers remain (avoid deleting valid numbers)
ds['Phone_Number'] = ds['Phone_Number'].apply(lambda x: x if len(x) == 10  else '')

# Step 3: Format as XXX-XXX-XXXX
ds['Phone_Number'] = ds['Phone_Number'].str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)

print('After Cleaning: ', ds['Phone_Number'])
```

## Splitting Address:

```python
ds[['Street', 'City', 'State_Zip']] = ds['Address'].str.split(',', n=2, expand=True)
ds

#ðŸ“Œ you can also use it when you have a name like Rida,Amelloul or Rida,Amelloul and you want to split it 
```

## Removing all of 'NaN', 'N/a',np.nan,'None' in datasets

```python
import numpy as np # type: ignore

columns = ['First_Name','Last_Name','Address','Paying Customer','Do_Not_Contact','Street','City','State_Zip']
for column in columns:
    ds[column] = ds[column].replace(['NaN', 'N/a',np.nan,'None'], '')
```

```python
ds
```

# Filling Yes:Y and No:N

```python
columns = ['Paying Customer','Do_Not_Contact']
for column in columns:
    ds[column] = ds[column].replace('Yes', 'Y')
    ds[column] = ds[column].replace('No', 'N')
```

## Removing Empty rows:

```python
for x in ds.index:
    if ds.loc[x, 'Phone_Number'] == '':
        ds.drop(x, inplace=True)
    elif (ds.loc[x, 'Do_Not_Contact'] == '') | (ds.loc[x, 'Do_Not_Contact'] == 'Y') :
        ds.drop(x, inplace=True)

ds
```
