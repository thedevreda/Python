```python
import pandas as pd
import matplotlib as plt
from datasets import load_dataset

#Importing datasets
ds = load_dataset("lukebarousse/data_jobs")
df = ds['train'].to_pandas()

#Cleaning datasets
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.index
```

## reset_index()

```python
# a function that create new index(id) and drop the old one 

# creating dataset and filtering y country from that dataset
df_morocco = df[df['job_country'] == 'Morocco' ] 

df_morocco.reset_index()
```

## sort_index()

```python
# is like sort_values() but for index

median = df_morocco.pivot_table(values=['salary_year_avg','salary_hour_avg'], index='job_title_short',aggfunc=['median','min','max'])

# you can say we can use sort values and put the name of index and sort by it butt....
median.sort_values(by=('median','salary_year_avg'), ascending=False)
# median.sort_values(by='job_title_short', ascending=False)

# what if we the index column doesn't have a  name
median.sort_index() # A-Z order
```
