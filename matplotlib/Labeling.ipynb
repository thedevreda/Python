```python
import pandas as pd
from datasets import load_dataset
from matplotlib import pyplot as plt

# Loading Dataset
datasets = load_dataset("lukebarousse/data_jobs")
df = datasets['train'].to_pandas()

# Cleaning Datasets
pd.to_datetime(df['job_posted_date'])

#Updating the dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df['job_posted_date'].dt.month
```

## Title (Old Way)

```python
job_count = df['job_title_short'].value_counts()
plt.bar(job_count.index, job_count.values)


# Title : 
plt.title("Posting by Job Title")

# X label :
plt.xlabel("Jobs Title")
# Y Label : 
plt.ylabel("Count of Job Posting")

# Customize X label Text
plt.xticks(rotation=45, ha='right')

plt.show()
```

## Title (New Way W/ Pandas):

```python
# With variable: 

job_count.plot(kind="bar", title="Posting by Job Title", xlabel="Jobs Title", ylabel="Count of Job Posting")
```

```python
# With Dataframe: 

df.plot(x="job_posted_month", y="salary_year_avg", kind="line")
```
