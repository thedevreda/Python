```python
import pandas as pd
from datasets import load_dataset
from matplotlib import pyplot as plt

# Loading Dataset
datasets = load_dataset("lukebarousse/data_jobs")
df = datasets['train'].to_pandas()

# Cleaning Datasets
pd.to_datetime(df['job_posted_date'])

#Updating the dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]
plt.plot(x, y)
plt.show() # to just get rid of this [<matplotlib.lines.Line2D at 0x28478cac9d0>]
```

## Count of Job Posted over Months

```python
# inserting month to avoid repating date and the time with seconds will be a bunch of miss
df['job_posted_month'] = df['job_posted_date'].dt.month

# sorting by index its gonna order it by values and the index will be a shit(5,12,3,0,120) values(120,150,245) 
# but sorting by index(1,2,3,4,5,6) and values(900,456,235)
# values count to count the values that reapted for each month
date_counts_monthly = df['job_posted_month'].value_counts().sort_index()

# x = date_counts_monthly.index (1,2,3,4,5,6)
# y = date_counts_monthly.values (900,123,1000,23)
plt.plot(date_counts_monthly.index, date_counts_monthly.values)
plt.show()
```
